## Strutured Programming

In modern programming languages (of course including Python and C), we can make programs with the concept of **structured programming.** Structured programming consists of the three elements.

- Sequence: A program run a program with the direction of **Top to bottom.**
- Selection: A program may have **a conditional execution statement**.
- Loop(Repetition, Iteration): A program may have **a repetition of executing statements**. Some parts of the program may be executed repeatedly.

Basically programmers must build programming flows by combinations of three elements above.

## Loop (Iteration, Repetition)
loop (iteration, repetition) means that it repeats certain parts of the program while satisfying certain conditions. 
In Python, there are two statements, `while` and `for` to represent a loop.

### (1) `while` statement 
Using `while` to represent a loop, write a program like:

(`cond` means a condition. You can use comparators and combinations as the same of `if` statement)
```
while cond:
  # A: you must indent when you indicate the statement is inside of the loop
  # B: do something  
else:
  # C: execute the statements when cond isn't satisfied.

# D: After the program exits the loop, then the statement is executed.
```
This program means:
- First, the program determines whether `cond` is satisfied. 
-- if `cond` is satisfied, then Python will execute `# A: ` and `# B: ` the indendted statements in order (top to bottom). 
Please be aware of indenting. When executing the indented statements are done, then Python goes back to the `while cond` statement and check `cond`  repeatedly.
-- if `cond` isn't satisfied, then Python executes `# C: `
-- Finally, Python executes `# D: `.

Please execute the following program and understand how the statements inside the loop are executed.

In this program, `c` is used for counter. 

`s` is used for storing the sum from 1 to `c`.

`c` should be incremented by 1 to calculate the sum.

In the program below, we used several `print()` to understand how the program behaves.
You can change `N` to various values, such as positive integer, 0 and negative integer. 

````python
# calculate the sum of 1 to N-1 by using loop.
N = 10 # change the value

c = 1
s = 0 # sum from 0 to c
while c <= N:
  print(str(c) + "th loop:")
  print("s is equal to", s)
  s += c # same as s = s + c
  print("s is changed to", s, "after executing s+=c, where c=",c)
  print("c is equal to", c)
  c += 1 # same as c = c + 1
  print("c is changed to", c, "after executing c+=1")
 
print("The sum of numbers from 1 to " + str(N) + " is :", s)
````

If you want to calculate the sum of **even numbers** from 1 to N, you can use `if` statement below.
**even number** means it is divisible by 2, so I used `if` statement to check the number.

- N = 0 -> sum = 0. In this case, the statements in the loop aren't executed.
- N = 1 -> sum = 0
- N = 3 -> sum = 2 (= 2) 
- N = 5 -> sum = 6 (= 2 + 4) 


In this program, we use a lot of `print()`. 
This is "print debbuging", a programmer or a tester can inspect executed statements and the value of the variable. 
It is often useful for debugging. 

(Debugging means removing mistakes, faults, failures in programs or systems) 

````python
# calculate the sum of even numbers from 1 to N-1 by using loop.
N = 10 # change the value

c = 1
s = 0 # sum from 0 to c
while c <= N:
  print(str(c) + "th loop:")
  if c % 2 == 0:
    print("s is equal to", s)
    s += c # same as s = s + c
    print("s is changed to", s, "after executing s+=c, where c=",c)
  print("c is equal to", c)
  c += 1 # same as c = c + 1
  print("c is changed to", c, "after executing c+=1")
 
print("The sum of even numbers from 1 to " + str(N) + " is :", s)
````

#### Nestings
You can nest `while` same as the `if` statement.

When you want to make a multiplication table, write a program below: 

````python
# Multiplication table ( x * y )
# In Thailand, a student learns 1x1 to 12x12

x = 1 # initialize x

while x <= 12:
  y = 1 # every time x changes, need to initialize y
  while y <= 12:
    print (x, "times", y, "is", x * y)
    y += 1
  x += 1
````

### (2) `for` statement

`for` statement is also used for loops.
In `for` statement Python extracts each element in order from a list or a tuple and executes statements inside of the loop. 
In Python, `for` statement is often used with `range()` function like this.

```
for element in range(0, 10):
  print(element)
```


````python
for element in range(0, 10):
  print(element)
````

`range()` function yields a sequence of numbers. `range()` function have three arguments at most.
- `range(stop_value)`: yields from 1 to `stop_value`-1 with an increase by 1. 
- `range(start_value, stop_value)` yields from `start_value` to `stop_value`-1 with an increase by 1.
- `range(start_value, stop_value, step_value)` yields from `start_value` to `stop_value`-1 with an increase by `step_value`.


The snippet (i.e. a part of) program is
```
for element in range(start_value, stop_value, step_value):
  print(element)
```


````python
start_value = 1
stop_value = 10
step_value = 2

for element in range(start_value, stop_value, step_value):
  print(element)
````

The program behaves the same as
```
element = start_value
while element < stop_value:
  print(element)
  element += step_value
```

````python
start_value = 1
stop_value = 10
step_value = 2

element = start_value
while element < stop_value:
  print(element)
  element += step_value
````


As a viewpoint of simplicity, using `for` statement is better than `while` if the sequence using in the loop can be generated by `range()`.

You can nest `for` like `while` and `if`.
